```python
import torch
from torchvision import datasets, transforms
from rich.console import Console
from rich.panel import Panel
from rich.pretty import Pretty

from kernels import get_kernel

# Rich terminal nesnesi
console = Console()

# 1. CUDA kernel'i al
activation = get_kernel("kernels-community/activation")

# 2. MNIST veri setinden bir Ã¶rnek al
transform = transforms.Compose([
    transforms.ToTensor(),                     # [0,1] arasÄ± normalizasyon
    transforms.Lambda(lambda x: x.view(-1))    # 28x28'i vektÃ¶re Ã§evir (784,)
])
mnist = datasets.MNIST(root="./data", train=False, download=True, transform=transform)
image_tensor, label = mnist[0]  # Ä°lk gÃ¶rÃ¼ntÃ¼yÃ¼ al
image_tensor = image_tensor.to(dtype=torch.float16, device="cuda")  # CUDA'ya gÃ¶nder

# 3. GELU aktivasyonu uygula
output_tensor = torch.empty_like(image_tensor)
gelu_kernel = activation.gelu_fast(output_tensor, image_tensor)

# 4. SonuÃ§larÄ± gÃ¶ster
console.rule("[bold green]MNIST GÃ¶rseli Ãœzerinde GELU Aktivasyonu")

console.print(Panel.fit("ğŸ–¼ï¸ [bold yellow]Input GÃ¶rsel Tensor (x)", style="bold"))
console.print(Pretty(image_tensor, expand_all=False))

console.print(Panel.fit("ğŸ¯ [bold green]Output Tensor (y) after GELU", style="bold green"))
console.print(Pretty(output_tensor, expand_all=False))

console.print(Panel.fit("ğŸ·ï¸ [bold cyan]Etiket (Label): " + str(label), style="bold cyan"))
console.print(Panel.fit("âš™ï¸ Kernel: gelu_fast", style="bold magenta"))
```
