```python
import torch
from rich.console import Console
from rich.table import Table
from rich.panel import Panel
from rich.pretty import Pretty

from kernels import get_kernel

console = Console()

# 1. Kernel'i indir
activation = get_kernel("kernels-community/activation")

# 2. CUDA'da rastgele tensor
x = torch.randn((10, 10), dtype=torch.float16, device="cuda")

# 3. Aktivasyon için boş tensor
y = torch.empty_like(x)

# 4. Özel GELU çekirdeğini çalıştır
gelu_kernel = activation.gelu_fast(y, x)

# 5. Sonuçları yazdır
console.rule("[bold green]CUDA GELU Aktivasyonu Sonuçları")

console.print(Panel.fit("🔢 [bold yellow]Input Tensor (x)", style="bold"))
console.print(Pretty(x, expand_all=True))

console.print(Panel.fit("🎯 [bold green]Output Tensor (y) after GELU", style="bold green"))
console.print(Pretty(y, expand_all=True))

console.print(Panel.fit("⚙️ [bold cyan]Kernel Function Info", style="bold cyan"))
console.print(gelu_kernel)
```

```python
import torch
from torchvision import datasets, transforms
from rich.console import Console
from rich.panel import Panel
from rich.pretty import Pretty

from kernels import get_kernel

# Rich terminal nesnesi
console = Console()

# 1. CUDA kernel'i al
activation = get_kernel("kernels-community/activation")

# 2. MNIST veri setinden bir örnek al
transform = transforms.Compose([
    transforms.ToTensor(),                     # [0,1] arası normalizasyon
    transforms.Lambda(lambda x: x.view(-1))    # 28x28'i vektöre çevir (784,)
])
mnist = datasets.MNIST(root="./data", train=False, download=True, transform=transform)
image_tensor, label = mnist[0]  # İlk görüntüyü al
image_tensor = image_tensor.to(dtype=torch.float16, device="cuda")  # CUDA'ya gönder

# 3. GELU aktivasyonu uygula
output_tensor = torch.empty_like(image_tensor)
gelu_kernel = activation.gelu_fast(output_tensor, image_tensor)

# 4. Sonuçları göster
console.rule("[bold green]MNIST Görseli Üzerinde GELU Aktivasyonu")

console.print(Panel.fit("🖼️ [bold yellow]Input Görsel Tensor (x)", style="bold"))
console.print(Pretty(image_tensor, expand_all=False))

console.print(Panel.fit("🎯 [bold green]Output Tensor (y) after GELU", style="bold green"))
console.print(Pretty(output_tensor, expand_all=False))

console.print(Panel.fit("🏷️ [bold cyan]Etiket (Label): " + str(label), style="bold cyan"))
console.print(Panel.fit("⚙️ Kernel: gelu_fast", style="bold magenta"))
```
